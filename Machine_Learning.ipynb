##<font color = green>Estilos de Aprendizado
Os estilos podem ser classificados em
- <b><font color = grey>prendizado supervisionado (supervised Learning)</b>

Dados de treinamento já conhecidos, rótulos pré definidos, conjunto de usuarios que já pré classificou, e a partir destes dados, você ira construir um modelo de classificação de previsão, o desafio é atingir um desempenho aceitavel

Existem varias abordagens que podem ser utilizados as principais delas são a Indução de Regras, Regressão, Técnica de Redes Neurais, Aprendizado em instâncias, Métodos Baseados em Kernels

- <b><font color = grey>Aprendizado não supervisionado (unsupervised Learning)</b>

Neste caso o dado não é rotulado e o resultado não é conhecido, significa que o  algoritimo vai tentar prever um resultado, as tecnicas utilizadas são de: Agrupamento de clustering, Detecção de anomalias, redes neurais geradoras adversárias

- <b><font color = grey>Aprendizado Semi SUpervisionado (semisupervised learning)</b>

Mistura dados rotulados e não rotulados, acontece quando é muito caro rotular um dado, exemplo rotulação de imagens. existe uma previsão desejavel, mas modelo precisa de estutura
Abordagens: Métodos Geradores, Separação de baixa densidade,. Métodos baseados em Grafos e Métodos heuristicos

- <b><font color = grey>Aprendizado por reforço (reinforcement learning)</b>

é uma prendizagem on o agente computacional interage com um ambiente dinâmico para atingir um objetivo, não possui dados rotulados, mas recebe feedback
Abordagens: Método de Monte Carlo, e método de referências temporais

##<font color = green>Indução de Hipótese e Viés Indutivo

<b>Indução de hipótese:</b>
Em aprendizagem de máquina(supervisionada), o objetivo é encontrar uma função que mapeie as entradas e saídas.

<b>Agente deve aprender</b> a função com base em alguns exemplos de entradas com os valores das saídas correspondetes.

Exemplo ou Instancia:
Formalmente, um exemplo é um par [x, f(x)]

*   Onde x é a entrada
*   e f(x) é a saida da função aplicada a x (conhecido tambem como target ou label)


###Indução:
dada uma coleção de exemplos, indução é uma maneira de encontrar uma função h que seja uma aproximação de [f]

###Hipótese:
a Função h é chamada de uma hipótese

###Generalização:
é a capacidade de uma função hipótese[h] prever corretamente exemplos ainda não vistos.

Exemplos de Indução:
*   Conjunto de exemplos usado na indução é chamado de conjunto de treinamento
*   a função hipótese[h] é consistente se ela concorda com [f] em todos os exemplos do conjunto de treinamento.

##Navalha de Ockham (ou principio da Parcimônia)

*   Maximize a combinação de <b><u>consistência e simplicidade.</b></u>
*   Ou seja, prefira a hipótese mais simples que seja consistente com os dados de treinamento.

##Overfitting e Underfitting

*   Se uma hipótese tem alta capacidade de previsão dos dados de treinamento e baixa capacidade de generalização, provavelmente o modelo pode ter sofrido overfitting
*   Se uma hipotótese tem baixa capacidade de previsão, mesmo nos dados de treinamento, pode ter sofrido undefitting;

##Viés Indutivo
Algoritmos possuem preferências quanto a representação dos dados e a geração de regras, que podem limitar a busca no estpaço de hipóteses.
